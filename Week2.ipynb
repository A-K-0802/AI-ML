Linear Regression

```python
#Write Code for Logistic Regression here
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from ucimlrepo import fetch_ucirepo 
breast_cancer_coimbra = fetch_ucirepo(id=451) 

x_train,x_test,y_train,y_test=train_test_split(np.array(breast_cancer_coimbra.data.features),np.array(breast_cancer_coimbra.data.targets),test_size=0.3)
model=LogisticRegression()
model.fit(x_train,y_train)

print(model.score(x_test,y_test))
```

SVM(Support Vector Machines)

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC 
from ucimlrepo import fetch_ucirepo 
breast_cancer_coimbra = fetch_ucirepo(id=451) 
clf=SVC(kernel='linear',C=4)
x_train,x_test,y_train,y_test=train_test_split(np.array(breast_cancer_coimbra.data.features),np.array(breast_cancer_coimbra.data.targets),test_size=0.3)
clf.fit(x_train,y_train)
print(clf.score(x_test,y_test))
```

# **LINEAR REGRESSION**

```python
#For more information for the dataser refer: https://archive.ics.uci.edu/dataset/477/real+estate+valuation+data+set
real_estate_dataset = pd.read_excel('/content/Real estate valuation data set.xlsx')
```

```python
real_estate_dataset.head()
```

```python
real_estate_dataset.tail()
```

```python
#Code for Linear Regression here
from sklearn.linear_model import LinearRegression
import matplotlib as plt
import numpy as np
import pandas as pd
from ucimlrepo import fetch_ucirepo 
real_estate_dataset = fetch_ucirepo(id=477) 
from sklearn.model_selection import train_test_split

model=LinearRegression()
X=real_estate_dataset.data.features
y=real_estate_dataset.data.targets

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)

model.fit(X_train,y_train)

y_predicted=model.predict(X_test)
print(y_predicted)
```
